```python
import pandas as pd
import matplotlib.pyplot as plt

#VGG16
df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Augmentation/VGG16/Default Augmentation/training_metrics.csv')
df2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Augmentation/VGG16/CLAHE Augmentation/training_metrics.csv')
df3 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Augmentation/VGG16/Ben Graham Augmentation/training_metrics.csv')
df4 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Augmentation/VGG16/CLAHE and Ben Augmentation/training_metrics.csv')
df5 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Augmentation/VGG16/Default and CLAHE Augmentation/training_metrics.csv')
print(df1.head())

# Plot settings
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Graph 1: Training and Validation Accuracy
axes[0].plot(df1['Epoch'], df1['Train_Accuracy'], label='Default Train Accuracy')
axes[0].plot(df1['Epoch'], df1['Val_Accuracy'], label='Default Val Accuracy')
axes[0].plot(df2['Epoch'], df2['Train_Accuracy'], label='CLAHE Train Accuracy')
axes[0].plot(df2['Epoch'], df2['Val_Accuracy'], label='CLAHE Val Accuracy')
axes[0].plot(df3['Epoch'], df3['Train_Accuracy'], label='Ben Graham Train Accuracy')
axes[0].plot(df3['Epoch'], df3['Val_Accuracy'], label='Ben Graham Val Accuracy')
axes[0].plot(df4['Epoch'], df4['Train_Accuracy'], label='CLAHE+Ben Train Accuracy')
axes[0].plot(df4['Epoch'], df4['Val_Accuracy'], label='CLAHE+Ben Val Accuracy')
axes[0].plot(df5['Epoch'], df5['Train_Accuracy'], label='Default+CLAHE Train Accuracy')
axes[0].plot(df5['Epoch'], df5['Val_Accuracy'], label='Default+CLAHE Val Accuracy')
axes[0].set_title('Training and Validation Accuracy')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(True)

# Graph 2: Training and Validation Loss
axes[1].plot(df1['Epoch'], df1['Train_Loss'], label='Default Train Loss')
axes[1].plot(df1['Epoch'], df1['Val_Loss'], label='Default Val Loss')
axes[1].plot(df2['Epoch'], df2['Train_Loss'], label='CLAHE Train Loss')
axes[1].plot(df2['Epoch'], df2['Val_Loss'], label='CLAHE Val Loss')
axes[1].plot(df3['Epoch'], df3['Train_Loss'], label='Ben Graham Train Loss')
axes[1].plot(df3['Epoch'], df3['Val_Loss'], label='Ben Graham Val Loss')
axes[1].plot(df4['Epoch'], df4['Train_Loss'], label='CLAHE+Ben Train Loss')
axes[1].plot(df4['Epoch'], df4['Val_Loss'], label='CLAHE+Ben Val Loss')
axes[1].plot(df5['Epoch'], df5['Train_Loss'], label='Default+CLAHE Train Loss')
axes[1].plot(df5['Epoch'], df5['Val_Loss'], label='Default+CLAHE Val Loss')
axes[1].set_title('Training and Validation Loss')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True)

# Adjust layout and save the figure
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Augmentation/Graphs/VGG16_training_metrics_summary.png')
plt.show()
```
