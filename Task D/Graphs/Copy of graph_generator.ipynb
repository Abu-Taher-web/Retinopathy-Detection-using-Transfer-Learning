```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Task D/Bagging/def_aug_unsample_DenseNet121_metrics.csv')
df2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Task D/Bagging/def_aug_unsampled_ResNet34_metrics.csv')
df3 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Task D/Bagging/def_aug_unsampleed_EfficientNet_metrics.csv')
df4 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Task D/Bagging/oversampled_DenseNet121_metrics.csv')
df5 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Task D/Bagging/oversampled_EfficientNet_metrics.csv')
df6 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Task D/Bagging/oversampled_ResNet34_metrics.csv')
print(df1.head(3))
print(df2.head(3))
print(df3.head(3))
print(df4.head(3))
print(df5.head(3))
print(df5.head(3))

# Plot settings
import matplotlib.pyplot as plt
import pandas as pd

# Load CSV files into dataframes
df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Task D/Bagging/def_aug_unsample_DenseNet121_metrics.csv')
df2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Task D/Bagging/def_aug_unsampled_ResNet34_metrics.csv')
df3 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Task D/Bagging/def_aug_unsampleed_EfficientNet_metrics.csv')
df4 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Task D/Bagging/oversampled_DenseNet121_metrics.csv')
df5 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Task D/Bagging/oversampled_EfficientNet_metrics.csv')
df6 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Task D/Bagging/oversampled_ResNet34_metrics.csv')

# Extract the epochs, accuracies, and losses
epochs = df1['Epoch']  # Assuming all CSVs have the same number of epochs

# Create a 1x2 figure
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# --- Plot Training and Validation Accuracy ---
axes[0].plot(epochs, df1['Train_Accuracy'], label='Train Accuracy - DenseNet121', color='blue', marker='o')
axes[0].plot(epochs, df1['Val_Accuracy'], label='Validation Accuracy - DenseNet121', color='orange', marker='x')

axes[0].plot(epochs, df2['Train_Accuracy'], label='Train Accuracy - ResNet34', color='green', marker='o')
axes[0].plot(epochs, df2['Val_Accuracy'], label='Validation Accuracy - ResNet34', color='red', marker='x')

axes[0].plot(epochs, df3['Train_Accuracy'], label='Train Accuracy - EfficientNet', color='purple', marker='o')
axes[0].plot(epochs, df3['Val_Accuracy'], label='Validation Accuracy - EfficientNet', color='brown', marker='x')

axes[0].plot(epochs, df4['Train_Accuracy'], label='Train Accuracy - Oversampled DenseNet121', color='cyan', marker='o')
axes[0].plot(epochs, df4['Val_Accuracy'], label='Validation Accuracy - Oversampled DenseNet121', color='magenta', marker='x')

axes[0].plot(epochs, df5['Train_Accuracy'], label='Train Accuracy - Oversampled EfficientNet', color='yellow', marker='o')
axes[0].plot(epochs, df5['Val_Accuracy'], label='Validation Accuracy - Oversampled EfficientNet', color='pink', marker='x')

axes[0].plot(epochs, df6['Train_Accuracy'], label='Train Accuracy - Oversampled ResNet34', color='grey', marker='o')
axes[0].plot(epochs, df6['Val_Accuracy'], label='Validation Accuracy - Oversampled ResNet34', color='black', marker='x')

# Set the title and labels for the accuracy plot
axes[0].set_title('Training and Validation Accuracy for All Models')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Accuracy')
axes[0].legend()

# --- Plot Training and Validation Loss ---
axes[1].plot(epochs, df1['Train_Loss'], label='Train Loss - DenseNet121', color='blue', marker='o')
axes[1].plot(epochs, df1['Val_Loss'], label='Validation Loss - DenseNet121', color='orange', marker='x')

axes[1].plot(epochs, df2['Train_Loss'], label='Train Loss - ResNet34', color='green', marker='o')
axes[1].plot(epochs, df2['Val_Loss'], label='Validation Loss - ResNet34', color='red', marker='x')

axes[1].plot(epochs, df3['Train_Loss'], label='Train Loss - EfficientNet', color='purple', marker='o')
axes[1].plot(epochs, df3['Val_Loss'], label='Validation Loss - EfficientNet', color='brown', marker='x')

axes[1].plot(epochs, df4['Train_Loss'], label='Train Loss - Oversampled DenseNet121', color='cyan', marker='o')
axes[1].plot(epochs, df4['Val_Loss'], label='Validation Loss - Oversampled DenseNet121', color='magenta', marker='x')

axes[1].plot(epochs, df5['Train_Loss'], label='Train Loss - Oversampled EfficientNet', color='yellow', marker='o')
axes[1].plot(epochs, df5['Val_Loss'], label='Validation Loss - Oversampled EfficientNet', color='pink', marker='x')

axes[1].plot(epochs, df6['Train_Loss'], label='Train Loss - Oversampled ResNet34', color='grey', marker='o')
axes[1].plot(epochs, df6['Val_Loss'], label='Validation Loss - Oversampled ResNet34', color='black', marker='x')

# Set the title and labels for the loss plot
axes[1].set_title('Training and Validation Loss for All Models')
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('Loss')
axes[1].legend()

# Adjust layout to prevent overlapping
plt.tight_layout()

# Save the figure to a file
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Task D/Graphs/training_validation_plots_all_models_begging.png')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Oversampling/Resnet34/training_metrics.csv')
print(df1.head())

# Plot settings
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Graph 1: Training and Validation Accuracy
axes[0].plot(df1['Epoch'], df1['Train_Accuracy'], label='Default Train Accuracy', marker = '.', markersize=12)
axes[0].plot(df1['Epoch'], df1['Val_Accuracy'], label='Default Val Accuracy', marker = '.', markersize=12)
axes[0].set_title('Training and Validation Accuracy')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('Accuracy')
axes[0].legend()
axes[0].grid(True)

# Graph 2: Training and Validation Loss
axes[1].plot(df1['Epoch'], df1['Train_Loss'], label='Default Train Loss', marker = '.', markersize=12)
axes[1].plot(df1['Epoch'], df1['Val_Loss'], label='Default Val Loss', marker = '.', markersize=12)
axes[1].set_title('Training and Validation Loss')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True)

# Adjust layout and save the figure
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Deep Learning Final Project Data Analysis/Oversampling/Graphs/Resnet34_training_metrics_summary_oversampling.png')
plt.show()
```
